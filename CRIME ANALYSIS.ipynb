CRIME ANALYSIS PEROORKADA

```python
# Filter the data for Peroorkada
peroorkada_data = df[df['police station'] == 'peroorkada']
unique_years_peroorkada = peroorkada_data['Year'].unique()
print(unique_years_peroorkada)
```

```python
import pandas as pd

# Filter data for Peroorkada police station
peroorkada_data = df[df['police station'] == 'peroorkada']

# Calculate total crimes for each year
total_crimes_per_year = peroorkada_data.groupby('Year').sum(numeric_only=True).sum(axis=1)

# Display total crimes per year
print("Total IPC crimes in each year (Peroorkada):")
print(total_crimes_per_year)
```

```python
import matplotlib.pyplot as plt

# Plot total crimes per year as a line chart
plt.figure(figsize=(8, 6))
plt.plot(total_crimes_per_year.index, total_crimes_per_year.values, marker='o', color='#3498DB', linestyle='-', linewidth=2, markersize=8)

# Add labels and title
plt.title("Total IPC Crimes Per Year in Peroorkada (2014-2020)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Crimes", fontsize=12)

# Display values on the line
for index, value in enumerate(total_crimes_per_year):
    plt.text(total_crimes_per_year.index[index], value + 2, str(value), ha='center', fontsize=10)

# Show gridlines for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
